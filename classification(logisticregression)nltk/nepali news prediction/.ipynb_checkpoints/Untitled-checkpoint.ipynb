```python
df=pd.read_csv('train.csv')
```

```python
df
```

```python
df.groupby('label').paras.count().plot.bar()
```

```python
words=stopwords.words('nepali')
# df['head']=df['headings'].apply(lambda x :" ".join([stemmer.stem(i) for i in re.sub("[।(),०-९<<?!,—,–,/,’,‘,:,\u200d]", " ",x).split() if i not in words]))
df['news']=df['paras'].apply(lambda x :" ".join([stemmer.stem(i) for i in re.sub("[।(),०-९<<?!,—,–,/,’,‘,:,\u200d]", " ",x).split() if i not in words]))
```

```python
df
```

```python
data=df[['label','news']]
data
```

```python
data['news'][1]
```

```python
data.dropna()
```

```python
x=data['news'].values
y=data['label'].values
vectorizer=TfidfVectorizer()
pipeline = Pipeline([('vect', vectorizer),
                     ('chi',  SelectKBest(chi2, k=1200)),
                     ('clf', LogisticRegression(random_state=0))])
model=pipeline.fit(x,y)
```

```python
cf
```

```python
x_test=cf['paras'].values
y_test=cf['label'].values
```

```python
print(classification_report(model.predict(x_test),y_test))
print(confusion_matrix(model.predict(x_test),y_test))
```

```python
print(accuracy_score(y_test,model.predict(x_test)))
```
