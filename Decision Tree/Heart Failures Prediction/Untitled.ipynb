```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import seaborn as sns
```

```python
df=pd.read_csv('heart_failure.csv')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
```

```python
df.corr()
```

```python
x=df[['age','anaemia','creatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','serum_sodium','smoking','time']]
y=df['DEATH_EVENT']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=34,test_size=0.20)
```

```python
tree=DecisionTreeClassifier(max_depth=200)
tree.fit(x_train,y_train)
pred=tree.predict(x_test)

```

```python
tree_accuracy= round(accuracy_score(y_test,pred), 5)*100 # Accuracy
tree_accuracy
```

```python
def impressions(model,accuracy):
    print('Accuracy: {} %'.format(accuracy))
    cm=confusion_matrix(y_test,pred)
    class_label = [0, 1]
    df_cm = pd.DataFrame(cm, index=class_label,columns=class_label)
    sns.heatmap(df_cm,annot=True,cmap='Set2',linewidths=2,fmt='d')
    plt.title("Confusion Matrix",fontsize=15)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()
```

```python
impressions(tree,tree_accuracy)
```
