```python
import pandas as pd
import re
```

```python
df=pd.read_csv("laptop_price.csv",encoding='latin1')
```

```python
df=df.drop('laptop_ID',axis=1)
```

```python
df.isnull().sum()
```

```python
df
```

```python
df['Ram']=[int(r.replace("GB",'')) for r in df['Ram']]
df['Memory'] = [int(re.sub(r'\D', '', m)) for m in df['Memory']]
df['Weight']=[int(re.sub(r'\D', '', m)) for m in df['Weight']]
```

```python
set(df['Company'])
```

```python
len(df.columns)
```

```python
set(df['TypeName'])
```

```python
names=pd.get_dummies(df['TypeName'],drop_first=False)
```

```python
df=pd.concat([df,names],axis=1)
df.columns
```

```python
df.corr()
```

```python
df=df.drop('TypeName',axis=1)
```

```python
data = df[['Inches', 'Ram', 'Memory', 'Weight', 'Price_euros', '2 in 1 Convertible', 'Gaming', 'Netbook', 'Notebook', 'Ultrabook', 'Workstation']]
```

```python
data
```

```python
x=data[['Inches', 'Ram', 'Memory', 'Weight', '2 in 1 Convertible', 'Gaming', 'Netbook', 'Notebook', 'Ultrabook', 'Workstation']]
y=data['Price_euros']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=30)
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python
f=x_test.values
g=y_test.values
print("the predicted values is",regression.predict(f[[45]]),"the real value is ",g[45])
```

```python
from sklearn.metrics import mean_squared_error

# Assuming regression is your regression model, x_test contains test features, and y_test contains true labels
predictions = regression.predict(x_test)

# Calculate mean squared error
mse = mean_squared_error(y_test, predictions)

# Print mean squared error
print("Mean Squared Error:", mse)
```

```python
from sklearn.metrics import r2_score

# Assuming regression is your regression model, x_test contains test features, and y_test contains true labels
predictions = regression.predict(x_test)

# Calculate R-squared score
r2 = r2_score(y_test, predictions)

# Print R-squared score
print("R-squared Score:", r2)
```
