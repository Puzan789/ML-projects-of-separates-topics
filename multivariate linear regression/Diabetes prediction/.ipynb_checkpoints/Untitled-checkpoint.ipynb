```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("diabetes.csv")
df
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(df.corr(),annot=True)
```

```python
df.corr()
```

```python
df_train,df_test=train_test_split(df,test_size=0.4,random_state=34)
```

```python
x=df_train.iloc[:,:-1]
y=df_train.iloc[:,-1:]
```

```python
regressor=LinearRegression()
regressor.fit(x,y)
```

```python
pred=df_test.iloc[123:124]
pred=pred.drop('Outcome',axis=1)
```

```python
regressor.predict(pred)
```

```python
y_test=df_test[['Outcome']].values
x_test=df_test.drop("Outcome",axis=1).values
```

```python
for i in range (len(y_test)):
    print("actual outcome",y_test[i][0],"predicted:",round(regressor.predict(x_test)[i][0]))
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,regressor.predict(x_test))
```

- evaluation shows that  means that 29.19% of the differences in the data points from their average can be understood or predicted by the factors (independent variables) in the model. The remaining 70.81% of the differences are not explained by the model and might be due to factors that were not considered or random chance.
