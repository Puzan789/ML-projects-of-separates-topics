```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
auto=df.iloc[:,1:]
```

```python
auto=auto.drop(['street','statezip','country'],axis=1)
```

```python
auto.corr()
```

```python
plt.figure(figsize=(20,6))
sns.heatmap(auto.corr(),annot=True)
plt.show()
```

```python
auto
```

```python
cities=pd.get_dummies(auto['city'],drop_first=False)
```

```python
auto=pd.concat([auto,cities],axis=1)
```

```python
# # plt.figure(figsize=())
# sns.heatmap(auto.corr(),annot=True,cmap='coolwarm')
# plt.show()
```

```python
auto=auto.drop('city',axis=1)
```

```python
df_train,df_test=train_test_split(auto,test_size=0.5,random_state=32)
```

```python
x_train=df_train.iloc[:,1:]
y_train=df_train['price']
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
pred=df_test[123:124]
price=pred['price']
pred=pred.drop('price',axis=1)
pred
```

```python
regressor.predict(pred)
```

```python
print("The predicted price is",regressor.predict(pred)[0],"Actual price is",price.values[0])
```

```python
price.values[0]
```

```python
from sklearn.metrics import r2_score
```

```python
X_test=df_test.iloc[:,1:]
y_test=df_test.iloc[:,0]
```

```python
r2_score(y_test,regressor.predict(X_test))
```
