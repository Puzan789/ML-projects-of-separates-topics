```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

- simple linear regression is a type of regression algorithms that models the relationship between single dependent and independent variables

```python
data=pd.read_csv("Salary_dataset.csv")
data.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
X = data[['YearsExperience']]
y = data['Salary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pre = regressor.predict(np.array([[5]]))
y_pre
```

```python
y_pred=regressor.predict(X_test)
```

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color="blue")
```

```python
plt.scatter(X_test,y_test,color='red')
plt.plot(X_test,regressor.predict(X_test),color="blue")
```

```python
y_train_pred = regressor.predict(X_train)

# Predictions on test data
y_test_pred = regressor.predict(X_test)
```

```python
# Calculate residuals for test data
test_residuals = y_test - y_test_pred

# Plotting residuals
plt.figure(figsize=(8, 6))
plt.scatter(X_test, test_residuals, color='orange', label='Residuals')
plt.axhline(y=0, color='red', linestyle='--', linewidth=2, label='Zero Residuals Line')
plt.xlabel('X')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.legend()
plt.show()
```
