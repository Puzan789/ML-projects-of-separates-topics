```python
from sklearn.model_selection import train_test_split
```

```python
x=df.iloc[:,0:-1]
y=df.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier=KNeighborsClassifier(n_neighbors=3).fit(x_train,y_train)
```

```python
print(classifier.score(x_train,y_train))
print(classifier.score(x_test,y_test))
```

```python
k=int(input("enter the value for k"))
train_accuracy=[]
test_accuracy=[]
for i in range(0,k):
    classifiers=KNeighborsClassifier(n_neighbors=i+1)
    classifiers.fit(x_train,y_train)
    train_accuracy.append(classifiers.score(x_train,y_train))
    test_accuracy.append(classifiers.score(x_test,y_test))
print(train_accuracy)
print(test_accuracy)
```

```python
for i in range(k):
  print("Epoch",i+1,"acc difference ",test_accuracy[i] - train_accuracy[i])
```

```python
plt.title("hellp")
neighbors = [i for i in range(1,k+1)]
plt.plot(neighbors,train_accuracy,label="Training accuracy")
plt.plot(neighbors,test_accuracy,label="Testing accuracy")
plt.legend()
plt.xlabel("Number of neighbours")
plt.ylabel("accuracy")
plt.show()
```

```python
classifier=KNeighborsClassifier(n_neighbors=8).fit(x_train,y_train)
```
